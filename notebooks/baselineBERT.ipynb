This file creates the baseline for our process in identifying NER for historical newspapers through fine-tuning a Transformer language model. The baseline model is created through running the same dataset (529 historical newspapers) using a baseline that says:
- About 7% of all words in texts are Named Entities.

This is drawn from the study conducted by Tolochko, Petro, Paul Balluff, Jana Bernhard, Sebastian Galyga, Noëlle S. Lebernegg, and Hajo G. Boomgaarden. 2024. “What’s in a Name? The Effect of Named Entities on Topic Modelling Interpretability.” <i>Communication Methods and Measures</i> 18 (4): 349–70. doi:10.1080/19312458.2024.2302120.

I will randomly select 7% of words and then randomly assign the selected word to one of the three Named Entities that I'm studying (Person, Organization, Location)

All code in this file are of my own.

```python
# Load dataset
import json
import random

storageJSONFilePath = "/Users/Jerry/Desktop/CS376/FinalProject/storageJSON.json"

with open(storageJSONFilePath, "r") as dataStorage:
    dataContent = json.load(dataStorage)

baselineDataLabel = {}

entitiesList = ["PERSON", "ORGANIZATION", "LOCATION"]

# Iterate through all data (newspaper pages) and randomly assign NER labels
for newspaperTitle, newspaperContent in dataContent.items():
    newspaperContentByWord = newspaperContent.split()
    newspaperWordCount = len(newspaperContentByWord)
    numberOfWordsNeededToSelect = int(newspaperWordCount * 0.07)
    randomlySelectedWords = random.sample(newspaperContentByWord, numberOfWordsNeededToSelect)
    baselineDataLabel[newspaperTitle] = []

    for word in randomlySelectedWords:
        randomIndex = random.randint(0, 2)
        baselineDataLabel[newspaperTitle].append([word, entitiesList[randomIndex]])
```

```python
baselineDataStorageJSONFilePath = "/Users/Jerry/Desktop/CS372/FinalProject/baselineLabelingStorageJSON.json"
with open(baselineDataStorageJSONFilePath, "w") as baselineStorage:
    json.dump(baselineDataLabel, baselineStorage)
```
